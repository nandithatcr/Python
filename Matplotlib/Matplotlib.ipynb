```python
import numpy as np
import matplotlib.pyplot as plt

ypoints = np.array([3, 8, 1, 10, 5, 7])

# Using fmt for formatting
plt.plot(ypoints, marker='o', linestyle=':', color='r',lw=2.5)  # Specify marker, linestyle, and color and line width
plt.show()
```

Change the size of marker using keyword marker size (short form ms)

```python
import numpy as np
import matplotlib.pyplot as plt

ypoints = np.array([3, 8, 1, 10, 5, 7])

# Using fmt for formatting
plt.plot(ypoints, marker='o',ms = 20)  # Specify marker, linestyle, and color
plt.show()
```

The subplot() function is used to create multiple plots in a single figure. It divides the figure into a grid of rows and columns and allows you to place one plot in each cell.
1) plt.subplot(nrows, ncols, index)

nrows: Number of rows in the grid.
ncols: Number of columns in the grid.
index: Index of the current subplot (1-based indexing).

# Matplotlib Scatter

Matplotlib's scatter() function is used to create scatter plots, which are useful for visualizing the relationship between two sets of data. Each point in a scatter plot represents an observation from the dataset.

```python
import matplotlib.pyplot as plt

# Data
x = [5, 7, 8, 7, 2, 17, 2, 9, 4, 11]
y = [99, 86, 87, 88, 100, 86, 103, 87, 94, 78]

# Scatter plot
plt.scatter(x,y)
plt.title("Basic Scatter Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

# Compare 2 plots

```python
x = np.array([5,7,8,7,2,17,2,9,4])
y=np.array([94,86,88,100,89,85,101,87,95])
plt.scatter(x,y)
x = np.array([2,3,8,1,15,8,12,0,7])
y=np.array([100,105,110,115,120,125,90,85,90])
plt.scatter(x,y)
plt.show()
```

Format the scatter points using argumments color size etc

```python
# Data
x = [1, 2, 3, 4, 5]
y = [10, 20, 25, 30, 35]

# Scatter plot with custom markers
plt.scatter(x, y, color='hotpink', marker='*', s=200, edgecolor='black')
plt.title("Scatter Plot with Custom Markers and other features")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

Color each dot by giving required colors as an array

```python
# Data
x = [1, 2, 3, 4, 5]
y = [10, 20, 25, 30, 35]

# Scatter plot with custom markers
colors = ['red','blue','green','black','yellow']
plt.scatter(x, y, c=colors)
plt.title("Scatter Plot with Custom Colors")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

# ColorMaps

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
x = np.random.rand(50)
y = np.random.rand(50)
colors = np.random.rand(50)

# Scatter plot with colormap
plt.scatter(x, y, c=colors, cmap='viridis', s=100, alpha=0.8)
plt.colorbar(label="Color Scale")
plt.title("Scatter Plot with Colormap")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 35, 15])
mylabels = ['Apples', 'Bananas', 'Cherries', 'Dates']
mycolors = ['black', 'hotpink', 'b', '#4CAF50']

plt.pie(y, labels=mylabels, colors=mycolors, autopct='%1.1f%%')  # Added percentages to the chart
plt.legend(loc='upper left', bbox_to_anchor=(1, 0.5))  # Moves the legend outside the chart
plt.tight_layout()  # Adjusts spacing to avoid overlap
plt.show()
```

# Wordcloud

What is a Word Cloud?
A word cloud (or tag cloud) is a visual representation of text data, where:

Words are displayed in varying sizes.
The size of each word is proportional to its frequency or importance in the text.
Word clouds are a popular data visualization tool, especially for textual data, as they provide a quick and intuitive way to understand the most prominent terms.

```python
from wordcloud import WordCloud
text = ("Python is an amazing programming language.It is widely used for data analysis, machine learning, and web development. Python oh oh oh oh verse wrote book stand title feel safe right away you dont get to decide")
Wordcloud = WordCloud(width=480,height=480,margin= 0).generate(text)
plt.imshow(Wordcloud,interpolation='bilinear')
plt.axis('off')
plt.margins(x=0,y=0)
plt.show()
```
